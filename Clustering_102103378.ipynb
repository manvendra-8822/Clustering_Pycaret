```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score

# Initialize an empty DataFrame to store the silhouette scores
no_pre = pd.DataFrame(columns=['Num Clusters', 'Silhouette Score'])

# Iterate over different numbers of clusters
for num_clusters in [3, 4, 5]:
    print(f"For Cluster = {num_clusters}")

    # Create the KMeans clustering model
    kmeans_model = create_model('kmeans', num_clusters=num_clusters)

    # Get cluster labels
    labels = kmeans_model.labels_

    # Compute silhouette score
    silhouette_score_value = silhouette_score(df, labels)
    calinski_harabasz_score_value = calinski_harabasz_score(df, labels)
    davies_bouldin_score_value=davies_bouldin_score(df,labels)
    # Store the silhouette score in the DataFrame
    no_pre = no_pre.append({'Num Clusters': num_clusters, 'Silhouette Score': silhouette_score_value, 'Calinski-Score':calinski_harabasz_score_value,'Davies-Bouldin':davies_bouldin_score_value}, ignore_index=True)
    new_no_pre=no_pre.set_index('Num Clusters')
# Print the silhouette scores DataFrame
print(new_no_pre)
```

```python
#NORMALIZATION

setup(data=df,normalize = True, normalize_method = 'zscore', verbose=False)

# Initialize an empty DataFrame to store the silhouette scores
normal = pd.DataFrame(columns=['Num Clusters', 'Silhouette Score'])

# Iterate over different numbers of clusters
for num_clusters in [3, 4, 5]:
    print(f"For Cluster = {num_clusters}")

    # Create the KMeans clustering model
    kmeans_model = create_model('kmeans', num_clusters=num_clusters)

    # Get cluster labels
    labels = kmeans_model.labels_

    # Compute silhouette score
    silhouette_score_value = silhouette_score(df, labels)
    calinski_harabasz_score_value = calinski_harabasz_score(df, labels)
    davies_bouldin_score_value=davies_bouldin_score(df,labels)
    # Store the silhouette score in the DataFrame
    normal = normal.append({'Num Clusters': num_clusters, 'Silhouette Score': silhouette_score_value, 'Calinski-Score':calinski_harabasz_score_value,'Davies-Bouldin':davies_bouldin_score_value}, ignore_index=True)
    new_normal=normal.set_index('Num Clusters')
# Print the silhouette scores DataFrame
print(new_normal)
```
